```python
import matplotlib.pyplot as plt
```

```python
import numpy as np 
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x**2)
ax.set(title="x^2")
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
dict = {"1": 10,
        "2": 8,
        "3": 12}
fig, ax = plt.subplots()
ax.barh(dict.keys(), dict.values())
```

```python
fig, ax = plt.subplots()
x = np.random.randn(10000)
ax.hist(x)
```
