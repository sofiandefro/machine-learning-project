```python
heart_disease = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv") # load directly from raw URL (source: https://github.com/mrdbourke/zero-to-mastery-ml/blob/master/data/heart-disease.csv)
heart_disease
```

```python
heart_disease["age"].hist(bins=50)
```

```python
heart_disease["trestbps"].hist(bins=50)
```

```python
plt.plot(np.linspace(0, 303, 303),heart_disease["trestbps"])
```

```python
heart_disease["chol"].hist(bins=50)
```

```python
heart_disease["fbs"].hist(bins=50)
```

```python
heart_disease["thalach"].hist(bins=50)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10,7))

scatter = ax[0].scatter(heart_disease["age"], heart_disease["chol"], c = heart_disease["target"], cmap='winter');
ax[0].set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol", xlabel="Age")
ax[0].legend(*scatter.legend_elements(), title="Target")
ax[0].axhline(y=heart_disease["chol"].mean(), color="r", linestyle="dashed", label="Average");


scatter = ax[1].scatter(heart_disease["age"], heart_disease["thalach"], c = heart_disease["target"], cmap='winter');
ax[1].set(title="Heart Disease and Max Heart Rate Levels", ylabel="Max Heart Rate", xlabel="Age")
ax[1].legend(*scatter.legend_elements(), title="Target")
ax[1].axhline(y=heart_disease["thalach"].mean(), color="r", linestyle="dashed", label="Average");

fig.suptitle("Heat Disease Analysis", fontsize=16, fontweight="bold");
```
